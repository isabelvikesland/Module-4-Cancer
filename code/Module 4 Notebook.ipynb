* Biological mechanisms (anatomy, organ physiology, cell & molecular physiology)

### Methods

*The machine learning techniques I am using are: Principal Component Analysis (PCA), UMAP, and K-Means Clustering.* Principal Component Analysis (PCA) is a dimensionality reduction technique that finds the directions of maximum variance in high-dimensional data and projects each sample onto those axes. PCA compresses the immense number of genes per person into a small number of "principal components" that capture the most important axes of transcriptomic variation. PC1 explains the most variance, PC2 the second most, and so on. By plotting PC1 vs. PC2, we get the best possible 2D linear summary of the full gene expression landscape across our 80 LUAD patients.

*What is PCA optimizing?* PCA finds the linear combination of genes that maximizes explained variance. Each successive principal component is orthogonal (uncorrelated) to all previous ones and explains as much remaining variance as possible. It does not use any clinical labels, it is purely driven by the structure of the expression data itself.

*UMAP (Uniform Manifold Approximation and Projection)* is a non-linear dimensionality reduction method that preserves local neighborhood structure: patients with similar overall gene expression profiles end up close together in the 2D projection. Unlike PCA, UMAP can capture curved or complex manifold structure that a linear projection would flatten out. The key hyperparameters are n_neighbors (how many nearby samples define a "neighborhood") and min_dist (how tightly samples are packed in the projection).

*K-Means Clustering* partitions patients into k non-overlapping groups by iteratively assigning each sample to its nearest cluster centroid and updating centroids until assignments stabilize. It is run on the top principal components (rather than all 15,716 raw genes) to reduce noise. The optimal number of clusters is chosen by maximizing the silhouette score, which measures how similar each sample is to its own cluster compared to neighboring clusters — a score of 1.0 means perfectly separated clusters, 0 means clusters are overlapping.

*What is K-Means optimizing?* K-Means minimizes within-cluster sum of squared distances, the total distance between each point and its cluster centroid. It stops when cluster assignments no longer change between iterations, or when a maximum iteration limit is reached.

## Load and subset data

```python
# Load expression matrix (genes x samples)
expr_raw = pd.read_csv('VALIDATION_SET_GSE62944_subsample_log2TPM.csv', index_col=0)
print(f'Expression matrix shape (genes x samples): {expr_raw.shape}')

# Load clinical metadata
meta_all = pd.read_csv('VALIDATION_SET_GSE62944_metadata.csv')
print(f'Metadata shape: {meta_all.shape}')
print(f'Cancer types: {meta_all["cancer_type"].value_counts().to_dict()}')

# Subset to LUAD only
luad_meta = meta_all[meta_all['cancer_type'] == 'LUAD'].copy().reset_index(drop=True)
luad_expr = expr_raw[luad_meta['sample'].values]   # genes x LUAD samples
print(f'LUAD expression shape (genes x samples): {luad_expr.shape}')
print(f'LUAD metadata shape: {luad_meta.shape}')

# Transpose: samples x genes (standard ML format)
X_all_genes = luad_expr.T   # 80 samples x 15716 genes
X_all_genes.index.name = 'sample'

print(f'\nTransposed expression matrix (samples x genes): {X_all_genes.shape}')

# Define the 9-gene metastasis signature from the paper
SIG_GENES = ['HMMR', 'B4GALT1', 'SLC16A3', 'ANGPTL4', 'EXT1', 'GPC1', 'RBCK1', 'SOD1', 'AGRN']

# Confirm all genes are present
missing = [g for g in SIG_GENES if g not in X_all_genes.columns]
if missing:
    print(f'WARNING — missing genes: {missing}')
else:
    print(f'All {len(SIG_GENES)} signature genes confirmed present.')

# Subset expression to signature genes only
X_sig = X_all_genes[SIG_GENES].copy()
print(f'Signature gene matrix shape: {X_sig.shape}')

# compute average signature expression score per patient (sample)
luad_meta = luad_meta.copy()
luad_meta.index = luad_meta['sample'].values
luad_meta['sig_score'] = X_sig.mean(axis=1).values

print(f'\nSignature score summary:')
print(luad_meta['sig_score'].describe().round(3))
```

## Compute score and clean data

```python
# clinical features for visualization

# Tumor stage: simplify to Stage I / II / III / IV
def simplify_stage(s):
    if pd.isna(s) or '[' in str(s):
        return 'Unknown'
    if 'IV' in s:
        return 'Stage IV'
    if 'III' in s:
        return 'Stage III'
    if 'II' in s:
        return 'Stage II'
    if 'I' in s:
        return 'Stage I'
    return 'Unknown'

luad_meta['stage_simple'] = luad_meta['ajcc_pathologic_tumor_stage'].apply(simplify_stage)
luad_meta['tumor_status_clean'] = luad_meta['tumor_status'].replace(
    {'[Not Available]': 'Unknown', '[Unknown]': 'Unknown'}
)

print('Stage distribution:')
print(luad_meta['stage_simple'].value_counts())
print('\nTumor status distribution:')
print(luad_meta['tumor_status_clean'].value_counts())
```

## PCA: Dimensionality Reduction

We apply PCA to the full gene expression matrix (all 15,716 genes), scaled to zero mean and unit variance. Using all genes captures global transcriptomic variation and places the 9-gene signature in broader biological context. We then color the PCA by:

1. Average signature score:does the metastasis signature drive variance?
2. Tumor stage: does transcriptomic variation correspond to clinical stage?
3. Tumor status: tumor free vs. with tumor

```python
#Scale and run PCA on all genes 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_all_genes)

pca = PCA(n_components=20, random_state=42)
X_pca = pca.fit_transform(X_scaled)

# Variance explained
var_exp = pca.explained_variance_ratio_ * 100
print('Variance explained by top 10 PCs:')
for i, v in enumerate(var_exp[:10]):
    print(f'  PC{i+1}: {v:.2f}%')
print(f'  Total (PC1-5): {var_exp[:5].sum():.2f}%')

#scree plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.bar(range(1, 21), var_exp, color='steelblue', edgecolor='white')
ax1.set_xlabel('Principal Component')
ax1.set_ylabel('Variance Explained (%)')
ax1.set_title('Scree Plot — Variance per PC')
ax1.set_xticks(range(1, 21))

ax2.plot(range(1, 21), np.cumsum(var_exp), 'o-', color='firebrick')
ax2.axhline(80, color='gray', linestyle='--', alpha=0.7, label='80% threshold')
ax2.set_xlabel('Number of PCs')
ax2.set_ylabel('Cumulative Variance Explained (%)')
ax2.set_title('Cumulative Variance Explained')
ax2.legend()
ax2.set_xticks(range(1, 21))

plt.suptitle('PCA Variance Explained — LUAD (all genes)', fontweight='bold')
plt.tight_layout()
plt.show()
```

## K-Means Clustering on PCA Space

We apply K-Means clustering to the top principal components (capturing the majority of variance) to ask: do LUAD patients naturally separate into transcriptomically distinct subgroups?

We use the silhouette score to select the optimal number of clusters k, then overlay cluster labels on the PCA plot.

```python
#Select number of PCs that explain >= 80% variance 
cumvar = np.cumsum(var_exp)
n_pcs = np.argmax(cumvar >= 80) + 1
print(f'Using top {n_pcs} PCs (explain {cumvar[n_pcs-1]:.1f}% of variance)')
X_pca_reduced = X_pca[:, :n_pcs]

# Silhouette score to choose optimal k for K-Means clustering on PCA-reduced data
sil_scores = []
k_range = range(2, 9)

for k in k_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_pca_reduced)
    sil = silhouette_score(X_pca_reduced, labels)
    sil_scores.append(sil)
    print(f'  k={k}: silhouette = {sil:.4f}')

best_k = list(k_range)[np.argmax(sil_scores)]
print(f'\nOptimal k = {best_k} (silhouette = {max(sil_scores):.4f})')

# Silhouette plot
plt.figure(figsize=(7, 4))
plt.plot(list(k_range), sil_scores, 'o-', color='steelblue', linewidth=2, markersize=8)
plt.axvline(best_k, color='firebrick', linestyle='--', label=f'Best k={best_k}')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs. k — K-Means on PCA Space')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Fit final K-Means with best k
km_final = KMeans(n_clusters=best_k, random_state=42, n_init=10)
cluster_labels = km_final.fit_predict(X_pca_reduced)
luad_meta['cluster'] = cluster_labels

cluster_palette = sns.color_palette('tab10', best_k)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Clusters on PCA
for c in range(best_k):
    mask = cluster_labels == c
    axes[0].scatter(pc1[mask], pc2[mask],
                   color=cluster_palette[c], label=f'Cluster {c} (n={mask.sum()})',
                   s=90, alpha=0.85, edgecolors='white', linewidth=0.5)
axes[0].set_xlabel(f'PC1 ({var_exp[0]:.1f}% variance)')
axes[0].set_ylabel(f'PC2 ({var_exp[1]:.1f}% variance)')
axes[0].set_title(f'K-Means Clusters (k={best_k}) on PCA')
axes[0].legend(fontsize=9)

#Clusters on UMAP
for c in range(best_k):
    mask = cluster_labels == c
    axes[1].scatter(u1[mask], u2[mask],
                   color=cluster_palette[c], label=f'Cluster {c}',
                   s=90, alpha=0.85, edgecolors='white', linewidth=0.5)
axes[1].set_xlabel('UMAP 1'); axes[1].set_ylabel('UMAP 2')
axes[1].set_title(f'K-Means Clusters (k={best_k}) on UMAP')
axes[1].legend(fontsize=9)

plt.suptitle(f'K-Means Clustering of LUAD Patients (k={best_k})',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

# characterize clusters by signature score and stage 
print('=== Cluster Characterization ===')
print('\nMean signature score by cluster:')
print(luad_meta.groupby('cluster')['sig_score'].agg(['mean','std','count']).round(3))

print('\nTumor stage distribution by cluster:')
print(pd.crosstab(luad_meta['cluster'], luad_meta['stage_simple']))

print('\nTumor status distribution by cluster:')
print(pd.crosstab(luad_meta['cluster'], luad_meta['tumor_status_clean']))
```
