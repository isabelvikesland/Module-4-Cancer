* Cancer Type: Lung Adenocarcinoma (LUAD)

We will focus solely on Lung Adenocarcinoma (LUAD), the most common subtype of non-small cell lung cancer (NSCLC). Lung adenocarcinoma is the most prevalent subtype of all lung cancers globally and accounts for approximately 40–50% of all NSCLC cases. The 5-year overall survival rate of LUAD remains at a low level of approximately 15.9%, largely because most patients are diagnosed at an advanced stage. In the United States, approximately 234,580 new lung cancer cases were projected for 2024, with an estimated 125,070 deaths — making lung cancer the leading cause of cancer-related mortality. https://pmc.ncbi.nlm.nih.gov/articles/PMC7183104/

* Prevalence & incidence

From 2000 to 2021, early-onset lung cancer incidence declined overall (average APC –3.81%; 95% CI, –4.01% to –3.64%; p < .0001). Globally in 2022, lung adenocarcinoma has emerged as the predominant subtype, with increasing incidence risks observed among younger generations,particularly among females. Adenocarcinoma now accounts for approximately 45.6% of male lung cancers and nearly 59.7% of female lung cancers worldwide. Incidence rates in the U.S. have been declining in men (~2.5% per year) but more slowly in women (~1% per year), largely reflecting historical smoking patterns. Mortality rates have declined by ~59% in men since 1990 and by ~36% in women since 2002, partly due to targeted therapy advances and earlier detection via low-dose CT (LDCT) screening.

* Risk factors (genetic, lifestyle) & Societal determinants

Cigarette smoking drives approximately 80% of all lung cancers, though LUAD is notably the subtype most common among never-smokers. Key driver mutations differ by smoking status:

Smokers: KRAS (especially G12C) and TP53 mutations predominate
Never-smokers: EGFR point mutations and gene fusions (ALK, ROS1, RET) are more prevalent

Genetic risk factors include germline EGFR pathogenic variants, BRCA2 variants (associated with ~12 years earlier onset), and Li-Fraumeni syndrome. Individuals with a first-degree relative with lung cancer face a ~50% increased risk (up to ~82% if a sibling). Environmental factors beyond smoking include radon gas (second leading cause of lung cancer in the U.S.), ambient air pollution (PM2.5), asbestos, and diesel exhaust. Notably, rising LUAD incidence among Asian and Hispanic females implicates emerging non-tobacco risk factors (e.g., air pollution, hormonal factors). Societal determinants are notable: lung cancer mortality is 4.6× higher among men with ≤12 years of education compared to college graduates, driven largely by smoking prevalence disparities.

* Standard of care treatments (& reimbursement) 

Treatment is now strongly driven by comprehensive molecular profiling via next-generation sequencing (NGS) for all advanced-stage LUAD patients. Key targeted therapies by driver mutation:

EGFR (exon 19 del, L858R): Osimertinib (3rd-generation TKI) — first-line standard
ALK fusions: Lorlatinib — 5-year PFS of 60% vs. 8% with crizotinib, the longest PFS reported for any single-agent targeted therapy in advanced solid tumors
KRAS G12C: Sotorasib, adagrasib — used post-immunotherapy/chemotherapy
ROS1, RET, MET exon 14, NTRK: Multiple approved TKIs available

For patients without actionable mutations, PD-1/PD-L1 immune checkpoint inhibitors (e.g., pembrolizumab, nivolumab, atezolizumab) form the treatment backbone, either as monotherapy (high PD-L1 expressors) or combined with platinum-based chemotherapy. Medical care coverage in the U.S. is generally available through Medicare, Medicaid, and private insurers for FDA-approved indications, though access to NGS profiling can be a barrier in lower-resource settings.

* Biological mechanisms (anatomy, organ physiology, cell & molecular physiology)

LUAD arises in the peripheral/distal lung, specifically the alveolar and bronchiolar epithelium, in contrast to squamous cell carcinoma, which originates in central airways. The cells lining the alveoli, alveolar type I (AT1) cells (responsible for gas exchange) and alveolar type II (AT2) cells (which produce surfactant and serve as stem-like progenitors), are the primary cells of origin for LUAD.

Under oncogenic stress (e.g., KRAS^G12D^ activation or carcinogen exposure), AT2 cells can dedifferentiate into a KRT8+ alveolar intermediate cell (KAC) state, a plastic, injury-responsive intermediate that has been identified as a key precursor to malignant transformation. LUAD follows a stepwise histological progression:

(1)Atypical adenomatous hyperplasia(AAH) (2)Adenocarcinoma in situ (AIS) (3)Minimally invasive adenocarcinoma(MIA) (4)Invasive adenocarcinoma

Histologically, LUAD growth patterns range from lepidic (low-grade; cells spread along existing alveolar walls) to acinar/papillary (intermediate-grade) to micropapillary/solid (high-grade). The proportion of high-grade components within a tumor is a key prognostic indicator.

RNA-sequencing (RNA-seq) was the technology used to generate the gene expression data. General overview of steps for RNA-sequencing:

(1)RNA is extracted from tumor tissue (fresh-frozen surgical specimens)
(2)RNA is reverse-transcribed to complementary DNA (cDNA)
(3)cDNA is fragmented, library-prepped, and sequenced on Illumina instruments
(4)Reads are aligned to the human reference genome (hg19) using STAR (2-pass mode)
(5)Transcript abundance is quantified using RSEM
(6)Values are normalized to TPM (transcripts per million), a measure of relative gene expression accounting for sequencing depth and gene length

Values are further log₂-transformed (log₂(TPM + 1)) to compress the wide dynamic range of expression data and approximate a normal distribution, which is important for downstream statistical analyses.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Loading Data
data = pd.read_csv(
    '/Users/tomas/CompBME/Module4_Vikesland+Daniel/Module-4-Cancer/data/TRAINING_SET_GSE62944_subsample_log2TPM.csv',
    index_col=0, header=0)
metadata_df = pd.read_csv(
    '/Users/tomas/CompBME/Module4_Vikesland+Daniel/Module-4-Cancer/data/TRAINING_SET_GSE62944_metadata.csv',
    index_col=0, header=0)

# Subsetting data before analysis
cancer_type = 'LUAD'
cancer_samples = metadata_df[metadata_df['cancer_type'] == cancer_type].index
LUAD_data = data[cancer_samples]

desired_gene_list = ['HMMR', 'B4GALT1', 'SLC16A3', 'ANGPTL4', 'EXT1', 'GPC1', 'RBCK1', 'SOD1', 'AGRN']
gene_list = [gene for gene in desired_gene_list if gene in LUAD_data.index]
for gene in desired_gene_list:
    if gene not in gene_list:
        print(f"Warning: {gene} not found in the dataset.")

LUAD_gene_data = LUAD_data.loc[gene_list]
LUAD_metadata = metadata_df.loc[cancer_samples]

# ── Prep matrix ───────────────────────────────────────────────────────────────
# Allows for figures to be scaled to display all data more effectively
X = LUAD_gene_data.T
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA Code
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

print(f"Explained variance ratio — PC1: {pca.explained_variance_ratio_[0]:.3f}, "
      f"PC2: {pca.explained_variance_ratio_[1]:.3f}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1],
                hue=LUAD_metadata['gender'],
                palette="Set2", s=100)
plt.title(f"PCA of {cancer_type} Gene Expression")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Gender")
plt.tight_layout()
plt.show()

# DBSCANS Code
dbscan = DBSCAN(eps=1.5, min_samples=5)
labels = dbscan.fit_predict(X_pca)

n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = (labels == -1).sum()
print(f"DBSCAN → {n_clusters} cluster(s), {n_noise} noise points (Teal colored points)")

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap="Set2", s=80)
plt.title(f"DBSCAN Clustering of {cancer_type}")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.tight_layout()
plt.show()
```
